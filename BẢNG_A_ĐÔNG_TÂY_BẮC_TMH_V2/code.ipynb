```python
# Của ảnh 9
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt

# Tham số mô hình
beta_11 = 0.4  # tỷ lệ lây nhiễm trong nhóm tuổi 1 (trẻ em)
beta_22 = 0.5  # tỷ lệ lây nhiễm trong nhóm tuổi 2 (người trưởng thành)
beta_33 = 0.4  # tỷ lệ lây nhiễm trong nhóm tuổi 3 (người già)
beta_12 = 0.6  # tỷ lệ lây nhiễm từ người trưởng thành sang trẻ em
beta_13 = 0.4  # tỷ lệ lây nhiễm từ người già sang trẻ em
beta_21 = 0.6  # tỷ lệ lây nhiễm từ trẻ em sang người trưởng thành
beta_23 = 0.7  # tỷ lệ lây nhiễm từ người già sang người trưởng thành
beta_31 = 0.6  # tỷ lệ lây nhiễm từ trẻ em sang người già
beta_32 = 0.7  # tỷ lệ lây nhiễm từ người trưởng thành sang người già

sigma_1 = 1/7  # tỷ lệ chuyển từ E sang I ở nhóm tuổi 1
sigma_2 = 1/3  # tỷ lệ chuyển từ E sang I ở nhóm tuổi 2
sigma_3 = 1/9  # tỷ lệ chuyển từ E sang I ở nhóm tuổi 3

gamma_1 = 1/10  # tỷ lệ hồi phục ở nhóm tuổi 1
gamma_2 = 1/7  # tỷ lệ hồi phục ở nhóm tuổi 2
gamma_3 = 1/17  # tỷ lệ hồi phục ở nhóm tuổi 3



# Tổng số dân cho mỗi nhóm tuổi
N1 = 10000  # số trẻ em
N2 = 10000  # số người trưởng thành
N3 = 10000  # số người già

# Điều kiện ban đầu
S1_0 = 6000  # số trẻ em nhạy cảm ban đầu
E1_0 = 2000     # số trẻ em đã tiếp xúc ban đầu
I1_0 = 2000     # số trẻ em bị nhiễm ban đầu
R1_0 = 0     # số trẻ em đã hồi phục ban đầu

S2_0 = 6000  # số người trưởng thành nhạy cảm ban đầu
E2_0 = 2000    # số người trưởng thành đã tiếp xúc ban đầu
I2_0 = 2000    # số người trưởng thành bị nhiễm ban đầu
R2_0 = 0    # số người trưởng thành đã hồi phục ban đầu

S3_0 = 6000  # số người già nhạy cảm ban đầu
E3_0 = 2000   # số người già đã tiếp xúc ban đầu
I3_0 = 2000   # số người già bị nhiễm ban đầu
R3_0 = 0    # số người già đã hồi phục ban đầu

# Hệ phương trình SEIR với nhiều nhóm tuổi
def seir_model_multi_age_no_birth(y, t, beta, sigma, gamma, N):
    S1, E1, I1, R1, S2, E2, I2, R2, S3, E3, I3, R3 = y
    dS1_dt = - (beta[0][0] * S1 * I1 / N[0] + beta[0][1] * S1 * I2 / N[0] + beta[0][2] * S1 * I3 / N[0])
    dE1_dt = (beta[0][0] * S1 * I1 / N[0] + beta[0][1] * S1 * I2 / N[0] + beta[0][2] * S1 * I3 / N[0]) - sigma[0] * E1
    dI1_dt = sigma[0] * E1 - gamma[0] * I1
    dR1_dt = gamma[0] * I1

    dS2_dt = - (beta[1][0] * S2 * I1 / N[1] + beta[1][1] * S2 * I2 / N[1] + beta[1][2] * S2 * I3 / N[1])
    dE2_dt = (beta[1][0] * S2 * I1 / N[1] + beta[1][1] * S2 * I2 / N[1] + beta[1][2] * S2 * I3 / N[1]) - sigma[1] * E2
    dI2_dt = sigma[1] * E2 - gamma[1] * I2
    dR2_dt = gamma[1] * I2

    dS3_dt = - (beta[2][0] * S3 * I1 / N[2] + beta[2][1] * S3 * I2 / N[2] + beta[2][2] * S3 * I3 / N[2])
    dE3_dt = (beta[2][0] * S3 * I1 / N[2] + beta[2][1] * S3 * I2 / N[2] + beta[2][2] * S3 * I3 / N[2]) - sigma[2] * E3
    dI3_dt = sigma[2] * E3 - gamma[2] * I3
    dR3_dt = gamma[2] * I3

    return dS1_dt, dE1_dt, dI1_dt, dR1_dt, dS2_dt, dE2_dt, dI2_dt, dR2_dt, dS3_dt, dE3_dt, dI3_dt, dR3_dt

# Thời gian mô phỏng
t = np.linspace(0, 50, 50)  # từ ngày 0 đến ngày 500

# Điều kiện ban đầu
y0 = S1_0, E1_0, I1_0, R1_0, S2_0, E2_0, I2_0, R2_0, S3_0, E3_0, I3_0, R3_0

# Tham số mô hình
beta = np.array([
    [beta_11, beta_12, beta_13],
    [beta_21, beta_22, beta_23],
    [beta_31, beta_32, beta_33]
])
sigma = np.array([sigma_1, sigma_2, sigma_3])
gamma = np.array([gamma_1, gamma_2, gamma_3])
N = np.array([N1, N2, N3])

# Giải hệ phương trình
solution = odeint(seir_model_multi_age_no_birth, y0, t, args=(beta, sigma, gamma, N))

# Trích xuất các giá trị
S1, E1, I1, R1, S2, E2, I2, R2, S3, E3, I3, R3 = solution.T

# Vẽ biểu đồ
plt.figure(figsize=(14, 10))

plt.subplot(3, 1, 1)
plt.plot(t, S1, label='Susceptible - Nhạy cảm')
plt.plot(t, E1, label='Exposed - Ủ bệnh')
plt.plot(t, I1, label='Infected - Phát bệnh')
plt.plot(t, R1, label='Recovered - Hồi phục')
plt.xlabel('Ngày')
plt.ylabel('Số dân')
plt.title('Trẻ em')
plt.legend()
plt.grid(True)

plt.subplot(3, 1, 2)
plt.plot(t, S2, label='Susceptible - Nhạy cảm')
plt.plot(t, E2, label='Exposed - Ủ bệnh')
plt.plot(t, I2, label='Infected - Phát bệnh')
plt.plot(t, R2, label='Recovered - Hồi phục')
plt.xlabel('Ngày')
plt.ylabel('Số dân')
plt.title('Người trưởng thành')
plt.legend()
plt.grid(True)

plt.subplot(3, 1, 3)
plt.plot(t, S3, label='Susceptible - Nhạy cảm')
plt.plot(t, E3, label='Exposed - Ủ bệnh')
plt.plot(t, I3, label='Infected - Phát bệnh')
plt.plot(t, R3, label='Recovered - Hồi phục')
plt.xlabel('Ngày')
plt.ylabel('Số dân')
plt.title('Người già')
plt.legend()
plt.grid(True)

plt.tight_layout()
plt.show()
```
